###  Importing necessary libraries ,mnist data set,keras and its layers ,models as squential,optimizer,performance metrics as accuracy

```python
import keras
from keras.datasets import mnist
from keras.layers import Conv2D,MaxPooling2D
from keras.layers import Dense,Flatten
from keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

```python
# Loading the dataset and perform splitting
(x_train,y_train),(x_test,y_test)=mnist.load_data()

### print the shape
print(x_train.shape)
print(x_test.shape)
print(y_train.shape[0])

# Peforming reshaping operation to conver image  to 1d array(28*28)
x_train=x_train.reshape(x_train.shape[0],28,28,1)
x_test=x_test.reshape(x_test.shape[0],28,28,1)

```

```python
# Normalization Each image has Intensity from 0 to 255
x_train = x_train / 255
x_test = x_test / 255

# 
# One Hot Encoding is required  for categorical feature since output to be predicted is handwritten image ,
### we have multiclass results(0â€“9: 10 output)
y_train =keras.utils.to_categorical(y_train,10)
y_test = keras.utils.to_categorical(y_test,10)
```

```python
model1 = Sequential([
    Conv2D(6, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(10, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(30, activation='relu'),
    Dense(10, activation='softmax')
])
print("--------- Model1 Summary----------")
print(model1.summary())
print('!!Starting  Trainging  on Model 1!')
# Compile and train the first model
model1.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
model1.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
print('Model 1 trained successfully!')




```

```python
model2 = Sequential([
    Conv2D(8, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(12, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(15, activation='relu'),
    Dense(10, activation='softmax')
])

print("--------- Model2 Summary---------")
print(model2.summary())
print('!!Starting  Trainging  on Model 2!')
# Compile and train the first model
model2.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
model2.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
print('Model 2 trained successfully!')

```

```python
model3 = Sequential([
    Conv2D(6, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(14, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(20,activation='relu'),
    Dense(20, activation='relu'),
    Dense(10, activation='softmax')
])

print("--------- Model3 Summary---------")
print(model3.summary())
print('!!Starting  Trainging  on Model 3!')
# Compile and train the first model
model3.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
model3.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
print('Model 3 trained successfully!')

```

```python
# Evaluate the models on the test set
_, accuracy1 = model1.evaluate(x_test, y_test, verbose=0)
_, accuracy2 = model2.evaluate(x_test, y_test, verbose=0)
_, accuracy3 = model3.evaluate(x_test, y_test, verbose=0)

# Create a comparison table
print ("Comparision Table of Various implemented CNN Architectures")
print(" CNN Architectures \t\t Accuracy")
print("-------------------------------------------------")
print("Model 1\t\t\t{:.2f}%".format(accuracy1 * 100))
print("Model 2\t\t\t{:.2f}%".format(accuracy2 * 100))
print("Model 3\t\t\t{:.2f}%".format(accuracy3 * 100))
```
